```python
import pandas as pd

df = pd.read_csv("./qualitaet_spritzguss.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Welche Spalten sollen zur Vorhersage verwendet werden
X = df[["Druck", "Temperatur"]].values
y = df["Erfolg"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='entropy', max_depth = 4)
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```

```python
from sklearn.tree import export_graphviz
tree = export_graphviz(model, None,
                      feature_names = ["Druck", "Temperatur"],
                      class_names = ["schlecht","gut"],
                      rounded = True,
                      filled = True)

import graphviz
graphviz.Source(tree)
#src = graphviz.Source(tree)
#src.render("Baum begrenzt")
```
