```python
import pandas as pd

df = pd.read_csv("./qualitaet_spritzguss.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Welche Spalten sollen zur Vorhersage verwendet werden
X = df[["Druck", "Temperatur"]].values
y = df["Erfolg"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.25)
```

```python
from sklearn.ensemble import AdaBoostClassifier

model = AdaBoostClassifier()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```

```python
# Hinweis: Damit dieser Befehl funktioniert, muss die 
# "helper.py" - Datei im selben Ordner liegen, wie das
# aktuelle Jupyter Notebook

from helper import plot_classifier

# Trainings-Daten plotten
plot_classifier(model, X_train, y_train, proba = False, 
                xlabel = "Druck", ylabel = "Temperatur",
                cmapName = "bwr")
```

```python
# Testdaten plotten

plot_classifier(model, X_test, y_test, proba = False, xlabel = "Druck", 
                ylabel = "Temperatur", cmapName = "bwr")
```

```python
# Plotten mit Wahrscheinlichkeiten
plot_classifier(model, X_test, y_test, proba = True, xlabel = "Druck", ylabel = "Temperatur")
```
